```python
ticker_BHP = yf.Ticker("BHP.AX")
data_BHP = ticker_BHP.history(start="2025-05-16", end="2025-05-17")

#We will be defining the Eurpoean Call Option parameters below

# Actual values
S0 = data_BHP["Close"].iloc[0] #Closing stock price on 16th May 2025
K = 0.98*S0 #98% of the stock price - this is the strike price of the option
T = 1 #T = 1 as this option will be for a year in the future
rf = yf.Ticker("^IRX") # This is the the 13 week Treasury Bill rate, which is used as the risk-free rate
data = rf.history(end="2025-05-16") # Get data before 16 May 2025
last_rf_rate = data["Close"].iloc[-1] / 100  # Convert from % to decimal
r = last_rf_rate  # Assigning the last risk-free rate to r
print(f"Last risk-free rate before 16 May 2025: {last_rf_rate:.4%}")

#Volatility calculation
BHP = yf.Ticker("BHP.AX")
data = BHP.history(start="2024-05-16", end="2025-05-16")  # 1 year of data
data['Daily Return'] = data['Close'].pct_change() # Calculate daily returns ib BHP
returns = data['Daily Return'].dropna()
daily_volatility = returns.std() # Standard deviation of daily returns
annual_volatility = daily_volatility * np.sqrt(252) # Annualised volatility
sigma = annual_volatility  # Assigning annual volatility to sigma
print(f"Annualized volatility for BHP: {annual_volatility:.4%}")
print(S0)

#To price the European Call Option, 3 different methods will be used.
#This is done to give a better understanding of different option pricing methods and will result in a more rigious pricing methodology.

#Option 1 - Black-Scholes Model

BHP_D1 = (np.log(S0 / K) + (r * T + 0.5 * sigma ** 2) * T) / (sigma * np.sqrt(T))
BHP_D2 = BHP_D1 - sigma * np.sqrt(T)

BHP_Call_Price = (S0 * norm.cdf(BHP_D1) - K * np.exp(-r * T) * norm.cdf(BHP_D2))

print("Black-Scholes Call Price: ", BHP_Call_Price)
```
